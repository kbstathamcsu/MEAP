```python
#Create a dataframe for each individual education level and LifeExp2018  and then calculate the correlation value
df_HSGradPer = df_merge[["hs_grad_or_higher", "LifeExp2018"]]
df_BADegPer = df_merge[["bs_degree_or_higher", "LifeExp2018"]]
df_AdvDegPer = df_merge[["adv_degree", "LifeExp2018"]]
corHS = df_HSGradPer.corr(method='pearson')
corBA = df_BADegPer.corr(method='pearson')
corAdv = df_AdvDegPer.corr(method='pearson')
corHSVal = corHS['LifeExp2018'][0].round(3)
corBAVal = corBA['LifeExp2018'][0].round(3)
corAdvVal = corAdv['LifeExp2018'][0].round(3)
```

```python
#Regression plot of each education level vs. LifeExp2018
fig, axes=plt.subplots(1,3,figsize=(12,8))
axes[0].set_title("Corr value = " + str(corHSVal))
axes[1].set_title("Corr value = " + str(corBAVal))
axes[2].set_title("Corr value = " + str(corAdvVal))
sns.regplot(x="hs_grad_or_higher", y="LifeExp2018", data=df_merge, label="hs_grad_or_higher", ax = axes[0], color="blue",
           ci=95)
sns.regplot(x="bs_degree_or_higher", y="LifeExp2018", data=df_merge, label="bs_degree_or_higher", ax = axes[1], 
            color = "orange", ci=95)
sns.regplot(x="adv_degree", y="LifeExp2018", data=df_merge, label="adv_degree", ax = axes[2], color = "gray", ci=95)
```

```python
#Create a dataframe for each type of crime and LifeExp2018  and then calculate the correlation value
df_MurderNMTot = df_merge[["Murders_NonNeg_Manslaughter 2015", "LifeExp2018"]]
df_MurderTot = df_merge[["Total Murders 2015", "LifeExp2018"]]
df_GunMurderTot = df_merge[["Total Gun Murders 2015", "LifeExp2018"]]
df_GunOwnPer = df_merge[["Gun Ownership 2013", "LifeExp2018"]]
corMurderNMTot = df_MurderNMTot.corr(method='pearson')
corMurderTot = df_MurderTot.corr(method='pearson')
corGunMurderTot = df_GunMurderTot.corr(method='pearson')
corGunOwnPer = df_GunOwnPer.corr(method='pearson')
corMurderNMTotVal = corMurderNMTot['LifeExp2018'][0].round(3)
corMurderTotVal = corMurderTot['LifeExp2018'][0].round(3)
corGunMurderTotVal = corGunMurderTot['LifeExp2018'][0].round(3)
corGunOwnPerVal = corGunOwnPer['LifeExp2018'][0].round(3)
```

```python
#Regression plot of murder types and gun ownership vs. LifeExp2018
fig, axes=plt.subplots(1,4,figsize=(12,8))
axes[0].set_title("Corr value = " + str(corMurderNMTotVal))
axes[1].set_title("Corr value = " + str(corMurderTotVal))
axes[2].set_title("Corr value = " + str(corGunMurderTotVal))
axes[3].set_title("Corr value = " + str(corGunOwnPerVal))

sns.regplot(x="Murders_NonNeg_Manslaughter 2015", y="LifeExp2018", data=df_merge, label="Murders_NonNeg_Manslaughter 2015", 
            ax = axes[0], ci=95, color="blue")
sns.regplot(x="Total Murders 2015", y="LifeExp2018", data=df_merge, label="Total Murders 2015", ci=95, ax=axes[1], color="orange")
sns.regplot(x="Total Gun Murders 2015", y="LifeExp2018", data=df_merge, label="Total Gun Murders 2015", ci=95, ax=axes[2],
           color="gray")
sns.regplot(x="Gun Ownership 2013", y="LifeExp2018", data=df_merge, label="Gun Ownership 2013", ci=95, ax=axes[3], color="purple")
```

```python
#Create a dataframe for each individual area characteristic and LifeExp2018  and then calculate the correlation value
df_TotalKmQ = df_merge[["TotalKmQ", "LifeExp2018"]]
df_LandKmQ = df_merge[["LandKmQ", "LifeExp2018"]]
df_WaterKmQ = df_merge[["WaterKmQ", "LifeExp2018"]]
corTotal = df_TotalKmQ.corr(method='pearson')
corLand = df_LandKmQ.corr(method='pearson')
corWater = df_WaterKmQ.corr(method='pearson')
corTotalVal = corTotal['LifeExp2018'][0].round(3)
corLandVal = corLand['LifeExp2018'][0].round(3)
corWaterVal = corWater['LifeExp2018'][0].round(3)
```

```python
#Regression plot of each land Characteristic vs. LifeExp2018
fig, axes=plt.subplots(1,3,figsize=(12,8))
axes[0].set_title("Corr value = " + str(corTotalVal))
axes[1].set_title("Corr value = " + str(corLandVal))
axes[2].set_title("Corr value = " + str(corWaterVal))
sns.regplot(x="TotalKmQ", y="LifeExp2018", data=df_merge, label="TotalKmQ", ax = axes[0], color="blue", ci=95)
sns.regplot(x="LandKmQ", y="LifeExp2018", data=df_merge, label="LandKmQ", ax = axes[1], color = "orange", ci=95)
sns.regplot(x="WaterKmQ", y="LifeExp2018", data=df_merge, label="WaterKmQ", ax = axes[2], color = "gray", ci=95)
```

```python
#Create a dataframe for the two income years and LifeExp2018  and then calculate the correlation value
df_Income2017 = df_merge[["Income2017", "LifeExp2018"]]
df_Income2010 = df_merge[["Income2010", "LifeExp2018"]]
cor2017 = df_Income2017.corr(method='pearson')
cor2010 = df_Income2010.corr(method='pearson')
cor2017Val = cor2017['LifeExp2018'][0].round(3)
cor2010Val = cor2010['LifeExp2018'][0].round(3)
```
